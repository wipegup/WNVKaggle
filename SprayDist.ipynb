```python
from scipy.spatial import cKDTree
import numpy as np

spray_dict = {}

for i in spray['Date'].dt.date.unique():
    
    spray_dict[i] = cKDTree( list(spray[spray['Date'] >= i ].loc[:,'Location']) )
```

```python
def sprayInstDict(sDics, trapDate, trapLoc, sprayDate):
    spray_inst = sDics[sprayDate].query(trapLoc, k = sDics[sDate].n)
    spray_inst = zip(spray_inst[0], spray_inst[1])
    toRet = {}
    
    for s in spray_inst:
        toRet[s[1]] = (s[0], trapDate - sprayDate)
    return toRet
```

```python
def findSprayDate(sprayDates, trapDate):
    sDates = sprayDates.sort(reverse = True)
    toRet = None
    for s in sprayDates:
        if (s.year == trapDate.year) & (s < trapDate):
            toRet = s
            break
    return toRet
```

```python
sorted_dates = list(spray_dict.keys())
sorted_dates.sort(reverse = True)

firstSpray = sorted_dates[-1]

trapSprays = {}
i=0

for l in df['Location'].unique():
    
    trapDates = df[ df['Location'] ==l]['Date'].unique()
    trapDates.sort()
    trapDates = [pd.Timestamp(d).date() for d in trapDates]
    
    if trapDates[-1] > firstSpray:
        trapSprays[l] = {}
        for tDate in trapDates:                
            sDate = findSprayDate(sorted_dates, tDate)
            
            if sDate:
                #print('hi')
                trapSprays[l][tDate] = sprayInstDict(spray_dict, tDate, l, sDate)
```
